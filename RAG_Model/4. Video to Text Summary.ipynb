```python
# # 100% working video --  short format
# https://www.youtube.com/watch?v=HbY51mVKrcE    ---- python example
import whisper

model = whisper.load_model("tiny", device="cpu")
result = model.transcribe("data/old.m4a", fp16=False)
#print(result["text"])
```

```python
print("Language: ",result['language'])
for i in result['segments']:
    print(i['id'],i['start'],"-->",i['end'],": ",i['text'])
    
```

```python
# # 100% working video
# https://www.youtube.com/watch?v=HbY51mVKrcE    ---- detailed python example for 30 sec chunks
import whisper

model = whisper.load_model("tiny")

# load audio and pad/trim it to fit 30 seconds
audio = whisper.load_audio("data/logan.mp3")
audio = whisper.pad_or_trim(audio)

# make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio, n_mels=model.dims.n_mels).to(model.device)

# detect the spoken language
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# decode the audio
options = whisper.DecodingOptions(fp16=False)
result = whisper.decode(model, mel, options)

# print the recognized text
print(result.text)
```
